## **5. Description of the dataset**

The .**describe()** function in pandas is a convenient method that provides summary statistics of a DataFrame or Series. It computes various descriptive statistics, such as count, mean, standard deviation, minimum, quartiles, and maximum, for the numerical columns in the DataFrame.

Here is a breakdown of the statistics provided by the `.describe()` function:

1. **Count**: The number of non-null values in each column.
2. **Mean**: The average value of each column.
3. **Standard Deviation (std)**: The measure of the amount of variation or dispersion in each column.
4. **Minimum (min)**: The smallest value in each column.
5. **25th Percentile (25%)**: Also known as the first quartile (Q1), this represents the value below which 25% of the data falls.
6. **50th Percentile (50%)**: Also known as the median, this represents the value below which 50% of the data falls.
7. **75th Percentile (75%)**: Also known as the third quartile (Q3), this represents the value below which 75% of the data falls.
8. **Maximum (max)**: The largest value in each column.

# **8.Normalize and Scale the dataset**

**Normalization:** It scales the values of a feature to a specific range, typically **between 0 and 1**. This ensures that features with different scales do not disproportionately influence the model's learning process. It's particularly useful for algorithms like k-nearest neighbors and neural networks that are sensitive to feature scaling.

**Scaling:** Scaling adjusts the distribution of data to have zero mean and unit variance. This helps algorithms that are sensitive to feature scaling, like support vector machines and linear regression.



# **Types of Scaling**
MinMaxScaler: This scaler transforms the data to a given range, usually between 0 and 1.

 Other Scalers:

  Scikit-learn provides other scaling methods like

  StandardScaler (for zero mean and unit variance) and RobustScaler (less sensitive to outliers).
  
  You can choose the most suitable scaler based on your data and model requirements

```python
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Apply the Min Mix Scaling to "Age" and "Scalary"
df_cleaned[['Age', 'Salary']] = scaler.fit_transform(df_cleaned[['Age', 'Salary']])

# Display the scaled data
df_cleaned.head()
```

# **8. Encode categorical Variables (Optional)**


Encoding Categorical Variables
Categorical variables are variables that can take on one of a limited, and usually fixed, number of possible values. Machine learning algorithms typically require **numerical inpu**t, so we need to encode these categorical variables into numerical representations.

There are two main types of encoding:

**1. One-Hot Encoding:**

What it does: Creates a new binary column for each category in the original column.
When to use: When the categorical variable has no inherent order or ranking.

**.2 Label Encoding:**

What it does: Assigns a unique numerical label to each category.
When to use: When the categorical variable has an inherent order or ranking (ordinal data).

# **Choosing the Right Encoding**

If your categorical variable has no inherent order, use One-Hot Encoding.

If your categorical variable has an inherent order, use Label Encoding.

Be careful of introducing unintended order with Label Encoding if the categories are not truly ordinal.
One-Hot Encoding can significantly increase the number of features, so consider dimensionality reduction techniques if necessary.

# **Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'Gender' is an ordinal categorical column
le = LabelEncoder()
df_cleaned['Gender_Encoded'] = le.fit_transform(df_cleaned['Gender'])

#Display encoder gender columns
df_cleaned[['Gender', 'Gender_Encoded']].head()
```

# **10. Detect outliers using Zero Score Scalling**

**Z-score scaling**, also known as **standardization**, is a statistical technique used to transform data into a distribution with a mean of 0 and a standard deviation of 1.

This transformation allows for the identification of outliers by measuring how many standard deviations a data point is away from the mean.

```python
from scipy import stats

import numpy as np


# Calculate Z-scores for 'Age' and 'Salary'
df_cleaned['Age_Zscore'] = stats.zscore(df_cleaned['Age'])
df_cleaned['Salary_Zscore'] = stats.zscore(df_cleaned['Salary'])

#identify rows with z- score > 3(outliers)
outliers = df_cleaned[(np.abs(df_cleaned['Age_Zscore']) > 3) | (np.abs(df_cleaned['Salary_Zscore']) > 3)]

#Rempve outliers from the dataset
df_cleaned = df_cleaned.drop(outliers.index)

# Display the cleaned datset without outliers
df_no_outliers = df_cleaned.drop(['Age_Zscore', 'Salary_Zscore'], axis=1)
df_no_outliers.head()
```
