# 1. Write a program that takes a string as input and counts the number of vowels and consonants in the string. (5 marks)

Input: Hello, World!

Output:

Number of vowels: 3

Number of consonants: 7

```python
def count_vowels_consonants(string):
  vowels = 0
  consonants = 0
  for char in string:
    if char.isalpha():
      if char.lower() in ['a','e','i','o','u']:
        vowels += 1
      else:
          consonants +=1
  return vowels,consonants

input_string= "Hello Word"
vowels,consonants = count_vowels_consonants(input_string)
print("Number of vowels:", vowels)
print("Number of consonants:", consonants)
```

# 6. Write a Python program that prompts the user to input a string. The program should count the frequency of each character in the string and then print the result as a dictionary.(5 marks)

EXAMPLE: If the user enters "hello" , the program should output {'h': 1, 'e': 1, 'l': 2, 'o': 1}
indicating that 'h' appears once, 'e' appears once, 'l' appears twice, and 'o' appears once in the input string.

```python
def count_char_freq(string):
  char_freq = {}
  for char in set(string):
    char_freq[char] = string.count(char)
    return char_freq

input_string = input("Enter a string:")
char_freq = count_char_freq(input_string)
print("Character frquency:", char_freq)
```

# 9. Create a list containing the 10 numbers . Print the first Three prime numbers from the list. (5 marks)
Expected Output:

List of the first three prime numbers: [2, 3, 5]

```python
#Create a list from1 to 10
numbers = list(range(1,11))
# Define a function to check if number is prime
def is_prime(num):
  if num <= 1:
    return False
  for i in range(2,int(num**0.5)+1):
    if num % i == 0:
     return False
  return True
#Find the first three prime number
prime_numbers = []
for num in numbers:
  if is_prime(num):
    prime_numbers.append(num)
    if len(prime_numbers) == 3:
     break
#Print the first three prime numbers
print("List of first three prime numers:", prime_numbers)
```

# 15.Write a program that takes a temperature in Celsius as input and converts it to Fahrenheit using the formula. (5 marks)

Fahrenheit = (Celsius * 9/5) + 32.

```python
def celsius_to_fahrenheit(celsius):
  """
  This function converts a temperature in Celsius to Fahrenheit.

  Args:
    celsius: The temperature in Celsius.

  Returns:
    The temperature in Fahrenheit.
  """
  fahrenheit = (celsius * 9/5) + 32
  return fahrenheit

# Take input from the user
celsius = float(input("Enter temperature in Celsius: "))

# Convert and print the result
fahrenheit = celsius_to_fahrenheit(celsius)
print(f"{celsius}°C is equal to {fahrenheit}°F")
```

```python
def reverse_list(list1):
  # Create an empty list to store the reversed elemnets
  reversed_list = []

  #iterate through the original list in reverse order
  for i in range(len(list1)-1,-1,-1):
      #Append each elemnet to the reversed list
     reversed_list.append(list1[i])

   # Return the reversed list
  return reversed_list

# Test the function
original_list = [1,2,3,4,5]
reversed_list = reverse_list(original_list)
print(reversed_list)
```

# 17.Given a string "Hello World", write a Python program to reverse the string without reversing the individual words. (5 marks)

Expected Output: “World Hello”

```python
def reverse_string_without_words(string):
    words = string.split()
    reversed_words = [word[::-1] for word in words]
    reversed_string = " ".join(reversed_words)
    return reversed_string
# Example use:
input_string = "Hello World"
output_string = reverse_string_without_words(input_string)
print("Reversed string without reversing individual words:", output_string)

# Output:
# Reversed string without reversing individual words: dlroW olleH
```

# 18. Implement a Python function is_palindrome() that checks whether a passed string is a palindrome (reads the same forward and backward) and tests it with "radar".(5 marks)

```python
def is_palindrome(string):
    reversed_string = ''.join(reversed(string))
    return string == reversed_string
# Output:
input_string = "radar"
is_palindrome_result = is_palindrome(input_string)
print("Is the string 'radar' a palindrome?", is_palindrome_result)


```

19.Write a program that returns a list of even numbers from a given list. (5 marks)

Input: [2, 5, 3, 7, 9, 53, 10, 32, 65, 76, 98]

```python
def even_numbers(numbers_list):
    return [num for num in numbers_list if num % 2 == 0]


# Output:
input_list = [2, 5, 3, 7, 9, 53, 10, 32, 65, 76, 98]
even_numbers_list = even_numbers(input_list)
print("List of even numbers:", even_numbers_list)
```

# 20.Given two lists keys = ['a', 'b', 'c'] and values = [1, 2, 3], write a Python program to create a dictionary from these lists. (5 marks)

Expected Output: {'a': 1, 'b': 2, 'c': 3}

```python
def create_dict_from_lists(keys, values):
  # Dictionary created using zip and dict
    dictionary = dict(zip(keys, values))
    return dictionary

keys = ['a', 'b', 'c']
values = [1, 2, 3]

result_dict = create_dict_from_lists(keys, values)
print("Dictionary created from keys and values:", result_dict)
```
